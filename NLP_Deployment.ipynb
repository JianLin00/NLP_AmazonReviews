```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora import Dictionary
from bertopic import BERTopic

# 2. Get documents from your DataFrame
docs = reviews_df["review"].dropna().astype(str).tolist()

# 3. Get topic words (excluding topic -1 if present)
topic_words = [
    [word for word, _ in topic_model.get_topic(topic_id)]
    for topic_id in topic_model.get_topic_freq().Topic
    if topic_id != -1
]

# 4. Prepare texts for gensim (tokenized)
texts = [doc.split() for doc in docs]

# 5. Create gensim dictionary and corpus
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# 6. Calculate coherence score
coherence_model = CoherenceModel(
    topics=topic_words,
    texts=texts,
    dictionary=dictionary,
    coherence='c_v'  # You can try 'u_mass' or others too
)

coherence_score = coherence_model.get_coherence()
print(f"BERTopic Coherence Score: {coherence_score:.3f}")
```

```python
# Diveristy

top_n = 10
all_words = []

for topic in topic_model.get_topics().values():
    words = [word for word, _ in topic[:top_n]]
    all_words.extend(words)

unique_words = set(all_words)
diversity_score = len(unique_words) / len(all_words)

print("Diversity Score:", diversity_score)
```

```python
import gradio as gr
import torch
import numpy as np
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
from bertopic import BERTopic

# Set device to CPU
device = torch.device("cpu")

# Load sentiment model from local folder
sentiment_model = DistilBertForSequenceClassification.from_pretrained("./sentiment_model").to(device)
sentiment_tokenizer = DistilBertTokenizer.from_pretrained("./sentiment_model")

# Load the previously saved BERTopic model
topic_model = BERTopic.load("bertopic_model")

# Define sentiment labels
sentiment_labels = ["Negative", "Neutral", "Positive"]

# Define the function to run both models
def analyze(text):
    try:
        if len(text.strip()) < 3:
            return "Too short", "-", "Too short"

        # Sentiment prediction
        inputs = sentiment_tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)
        outputs = sentiment_model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        confidence, predicted_label = torch.max(probs, dim=1)
        sentiment = sentiment_labels[predicted_label.item()]
        confidence_percent = f"{confidence.item() * 100:.2f}%"

        # Topic modeling
        topics, probs = topic_model.transform([text])
        topic_idx = topics[0]

        if topic_idx == -1 and probs is not None:
          # Fallback to highest-probability topic if -1 is assigned
          topic_idx = int(np.argmax(probs[0]))

        if topic_idx == -1:
          topic_output = "No strong topic match found."
        else:
          topic_words = topic_model.get_topic(topic_idx)
          if topic_words:
            topic_output = f"Topic {topic_idx}: " + ", ".join([word for word, _ in topic_words[:5]])
          else:
            # This case should ideally not happen for a valid topic_idx
            topic_output = f"Topic {topic_idx}: Information not available."


        return sentiment, confidence_percent, topic_output

    except Exception as e:
        return "Error", "-", f"Exception: {str(e)}"

# Gradio UI
with gr.Blocks(theme=gr.themes.Base(), css=".gr-button { font-size: 16px }") as demo:
    gr.Markdown("## 🧠 Sentiment & Topic Classification App")
    gr.Markdown("Enter a text to classify both its **sentiment** and **topic**.")

    with gr.Row():
        with gr.Column(scale=1):
            input_text = gr.Textbox(label="Input Text", lines=5, placeholder="Enter review or feedback...")
            analyze_btn = gr.Button("Analyze", variant="primary")
        with gr.Column(scale=1):
            sentiment_output = gr.Label(label="Predicted Sentiment")
            confidence_output = gr.Label(label="Confidence")
            topic_output = gr.Textbox(label="Top Topic (from BERTopic)", interactive=False)

    analyze_btn.click(fn=analyze, inputs=input_text, outputs=[sentiment_output, confidence_output, topic_output])
```

```python
import gradio as gr
import torch
import numpy as np
import pandas as pd
import os
import requests
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
from bertopic.backend import BaseEmbedder
from datetime import datetime
import smtplib
from email.message import EmailMessage


SENDER_EMAIL = "teokaining1130@gmail.com"
APP_PASSWORD =  "xxxxxxx" # from Gmail App Passwords, not your account password

def send_email(subject, body, to_email):
    msg = EmailMessage()
    msg.set_content(body)
    msg["Subject"] = subject
    msg["From"] = SENDER_EMAIL
    msg["To"] = to_email
    try:
        with smtplib.SMTP_SSL("smtp.gmail.com", 465) as smtp:
            smtp.login(SENDER_EMAIL, APP_PASSWORD)
            smtp.send_message(msg)
    except Exception as e:
        print(f"❌ Email error: {e}")

# -------- Topic Mapping --------
custom_topic_labels = {
    0: "Mobile Storage Cards", 1: "GoPro Action Cameras", 2: "General Product Feedback",
    3: "Microsoft Surface Tablets", 4: "Dash Cameras", 5: "Samsung Galaxy Phones",
    6: "SanDisk Branding and Quality", 7: "Memory Capacity Needs", 8: "Performance and Ease of Use",
    9: "File System Compatibility", 10: "Media Consumption on Tablets", 11: "SIM and Memory Chip Performance",
    12: "Smartphone Satisfaction", 13: "Dell Venue Devices", 14: "Shipping Speed and Packaging",
    15: "Samsung Tablets (incl. International)", 16: "Read/Write Speed", 17: "Android Tablet Installation",
    18: "Barnes & Noble Nook Tablets", 19: "General Memory Card Feedback",
    20: "MicroSD Pricing & Retail Experience", 21: "Storage Capacity Value",
    22: "GPS & Navigation Devices", 23: "Camera Usage for Photos", 24: "Data Backup and Transfers",
    25: "Device Formatting Requirements", 26: "ASUS Transformer Tablets",
    27: "Card Unmounting Problems", 28: "MP3 Players (Sansa Clip)", 29: "Pricing & Value for Phones",
    30: "MicroSD Packaging & Issues", -1: "Uncategorized"
}

# -------- Load Models --------
class SentenceTransformerEmbedder(BaseEmbedder):
    def __init__(self, model_name):
        self.model = SentenceTransformer(model_name)

    def embed(self, documents, verbose=False):
        return self.model.encode(documents, show_progress_bar=verbose)

embedding_model = SentenceTransformerEmbedder("all-mpnet-base-v2")
topic_model = BERTopic.load("bertopic_model_tuned")
topic_model.embedding_model = embedding_model

device = torch.device("cpu")
sentiment_model = DistilBertForSequenceClassification.from_pretrained("./sentiment_model").to(device)
sentiment_tokenizer = DistilBertTokenizer.from_pretrained("./sentiment_model")
sentiment_labels = ["Negative", "Neutral", "Positive"]

# -------- Analyze Function --------
def analyze(text, rating, tags, email, send_copy):
    try:
        if len(text.strip()) < 3:
            return "⚠️ Please enter a longer review.", None

        inputs = sentiment_tokenizer(text, return_tensors="pt", truncation=True, padding=True)
        outputs = sentiment_model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        confidence, predicted_label = torch.max(probs, dim=1)
        sentiment = sentiment_labels[predicted_label.item()]
        confidence_percent = f"{confidence.item() * 100:.2f}%"


        # Topic modeling
        topics, topic_probs = topic_model.transform([text])
        topic_idx = topics[0]

        if topic_idx == -1 and topic_probs is not None:
            topic_idx = int(np.argmax(topic_probs[0]))

        if topic_idx == -1:
            topic_output = "No strong topic match found."
        else:
            topic_words = topic_model.get_topic(topic_idx)
            if topic_words:
                topic_output = f"Topic {topic_idx}: " + ", ".join([word for word, _ in topic_words[:5]])
            else:
                topic_output = f"Topic {topic_idx}: No topic words available."

        if sentiment == "Positive":
            feedback = "😊 Thank you! We're glad you're happy with us!"
        elif sentiment == "Negative":
            feedback = "🙁 We're sorry to hear that. We'll work to improve!"
        else:
            feedback = "😐 Thanks for your review. We appreciate your input."

        now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        log_data = {
            "Date": now,
            "Review": text,
            "Sentiment": sentiment,
            "Confidence": confidence_percent,
            "Topic": topic_output,
            "Rating": rating,
            "Tags": ";".join(tags) if tags else "",
            "Email": email
        }
        log_file = "reviews_log.csv"
        if os.path.exists(log_file):
            df = pd.read_csv(log_file)
            df = pd.concat([df, pd.DataFrame([log_data])], ignore_index=True)
        else:
            df = pd.DataFrame([log_data])
        df.to_csv(log_file, index=False)

        summary = f"""
Review Summary
--------------
Date: {now}
Rating: {rating}
Tags: {', '.join(tags) if tags else 'None'}
Review: {text}
Feedback: {feedback}
"""

        if send_copy and email:
            send_email("Thank you for your feedback!", summary, email)

        with open("feedback_summary.txt", "w") as f:
            f.write(summary.strip())

        return f"""
✅ **Thank you for your feedback!**

📅 {now}
⭐ Rating: {rating}/5
🔖 Tags: {', '.join(tags) if tags else 'None'}

{feedback}
""", "feedback_summary.txt"

    except Exception as e:
        return f"❌ Error: {str(e)}", None

# -------- Gradio UI --------
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("## 📝 Customer Feedback Form")
    gr.Markdown("We’d love to hear your thoughts! Help us improve.")

    with gr.Row():
        input_text = gr.Textbox(label="Your Review", lines=5, placeholder="Share your experience...")
    with gr.Row():
        rating = gr.Slider(1, 5, step=1, label="Your Rating (1–5 stars)")
        tags = gr.CheckboxGroup(["Delivery", "Price", "Support", "Product"], label="What can we improve?")
    with gr.Row():
        email = gr.Textbox(label="Your Email (optional)", placeholder="you@example.com")
        send_copy = gr.Checkbox(label="Send me a copy by email")
    with gr.Row():
        analyze_btn = gr.Button("🚀 Submit Feedback")
    with gr.Row():
        feedback_output = gr.Markdown(label="📬 Our Response")
        download = gr.File(label="📥 Download Summary")

    analyze_btn.click(fn=analyze, inputs=[input_text, rating, tags, email, send_copy], outputs=[feedback_output, download])
```
